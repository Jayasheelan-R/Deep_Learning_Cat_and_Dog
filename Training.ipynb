```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense,Dropout,Activation,Flatten
import pickle
from keras.layers import Conv2D,MaxPooling2D
```

```python
pickle_in=open("E:\EVILDOLL\PYTHON PROJECTS\TENSER FLOW\Deep learning Hacker Earth\Cat and Dog\PetImages\X.pickle","rb")
X=pickle.load(pickle_in)

pickle_in=open("E:\EVILDOLL\PYTHON PROJECTS\TENSER FLOW\Deep learning Hacker Earth\Cat and Dog\PetImages\y.pickle","rb")
y=pickle.load(pickle_in)
print(X)
print(y)

X=X/255.0

print(X)
```

```python
model= Sequential()

model.add(Conv2D(256,(3,3),input_shape=X.shape[1:]))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(256,(3,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())
model.add(Dense(64))
model.add(Dense(1))
model.add(Activation('sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X,y,batch_size=50,epochs=1,validation_split=0.3)
```
